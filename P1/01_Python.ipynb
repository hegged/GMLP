<b>Exercise 4:</b> Word Frequency Counter

<b>Task:</b> 
Write a function word_frequency(text) that takes a string text as input and returns a dictionary where the keys are words and the values are the number of times each word appears in the text.
Words should be treated as case-insensitive (e.g., "The" and "the" are considered the same word).
Exclude punctuation from consideration (e.g., "hello," and "hello" are the same word).

Use the text: "To be, or not to be, that is the question:
Whether it is nobler in the mind to suffer
The slings and arrows of outrageous fortune,
Or to take arms against a sea of troubles
And by opposing end them."

<b>Guidelines:</b>
Consider using Python's built-in str methods, such as .lower() for case normalization and .translate() for punctuation removal.
The collections module, specifically Counter, can be very helpful for the word frequency counter task.
For cleaning the text, you may find the string module useful, especially string.punctuation for identifying characters to exclude.

```python
def word_frequency(text):
    import string
    from collections import Counter

    # Normalize the text to lowercase
    text = text.lower()

    # Remove punctuation
    translator = str.maketrans('', '', string.punctuation)
    cleaned_text = text.translate(translator)

    # Split the cleaned text into words
    words = cleaned_text.split()

    # Count the frequency of each word
    frequency = Counter(words)

    return dict(frequency)
word_frequency("""To be, or not to be, that is the question:
Whether it is nobler in the mind to suffer The slings and arrows of outrageous fortune,
Or to take arms against a sea of troubles And by opposing end them.""")
```
