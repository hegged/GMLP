Load the Iris dataset from Scikit-learn, convert it to a Pandas DataFrame, and use Seaborn to create a pair plot showing the pairwise relationships between the features, colored by species.
<ol>

  <li>Convert the data to a Pandas DataFrame (columns should be the feature names).</li>
  <li>Add a column 'species' (target) where the numeric target is exchanged by target names (e.g., 'setosa'). "Categorical.from_codes" could be helpfull</li>

```python
import seaborn as sea
import matplotlib.pyplot as plt
import pandas as pd
from seaborn import histplot
from sklearn import datasets

iris = datasets.load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['species'] = iris.target_names[iris.target]
sea.pairplot(df, hue='species')
plt.show()
histplot(data=df, x='sepal length (cm)', hue='species', kde=True)
```

```python
sea.boxplot(x='species', y='sepal length (cm)', data=df)
plt.show()
sea.violinplot(x='species', y='sepal length (cm)', data=df)
plt.show()
```
