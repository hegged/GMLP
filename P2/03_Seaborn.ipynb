<H1>Assignment 3: Seaborn</H1>

Seaborn is a Python data visualization library based on matplotlib that provides a high-level interface for drawing attractive and informative statistical graphics. In this assignment, we'll explore the Seaborn library using datasets available from Scikit-learn, converting them into Pandas DataFrame for easier manipulation and visualization. Make sure you have Seaborn, Matplotlib, Pandas, and Scikit-learn installed in your environment before proceeding.

<H2>Tools and Libraries</H2>
